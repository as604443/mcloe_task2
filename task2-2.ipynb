```python
mean_rd_spend = data['R&D Spend'].mean()
mean_administration = data['Administration'].mean()

data['R&D Spend'].fillna(mean_rd_spend, inplace=True)
data['Administration'].fillna(mean_administration, inplace=True)
```

```python
data
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(data['R&D Spend'], kde=True)
plt.subplot(2, 2, 2)
sns.histplot(data['Administration'], kde=True)
plt.subplot(2, 2, 3)
sns.histplot(data['Marketing Spend'], kde=True)
plt.subplot(2, 2, 4)
sns.histplot(data['Profit'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.scatterplot(x='R&D Spend', y='Profit', data=data)
plt.subplot(1, 3, 2)
sns.scatterplot(x='Administration', y='Profit', data=data)
plt.subplot(1, 3, 3)
sns.scatterplot(x='Marketing Spend', y='Profit', data=data)
plt.show()
```

```python
data = pd.get_dummies(data, columns=['State'], drop_first=True)
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
sns.countplot(x='State', data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(x=data['R&D Spend'])
plt.subplot(2, 2, 2)
sns.boxplot(x=data['Administration'])
plt.subplot(2, 2, 3)
sns.boxplot(x=data['Marketing Spend'])
plt.subplot(2, 2, 4)
sns.boxplot(x=data['Profit'])
plt.show()
```

```python

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((data < lower_bound) | (data > upper_bound)).any(axis=1)
data_no_outliers = data[~outliers]
data_no_outliers.to_csv("cleaned_data.csv", index=False)
```
